```python
import pandas as pd
import numpy as np
```

```python
# Load original data
df = pd.read_csv(r'data/fifa21_male2.csv',
                dtype={'Hits': 'object'})
df.info(verbose=True)
```

```python
# Standardize columns' names and change index
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_').str.replace('&', 'and')
if df.id.is_unique:
    df = df.set_index('id')

df.head()
```
